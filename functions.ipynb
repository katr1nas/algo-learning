```python
import numpy as np
import matplotlib.pyplot as plt

def generate_wn(quantity):
  e_values = []
  for t in range(quantity):
    e = np.random.randn()
    e_values.append(e)
  return e_values

wn = generate_wn(1000)
plt.plot(wn)
plt.show()
```

```python
def generate_wn(n , generator_type):
  e_values = []
  for t in range(n):
    if generator_type == 'U':
      e = np.random.uniform(0, 1)
    else:
      e = np.random.randn()
    e_values.append(e)

  return e_values

wn = generate_wn(100, 'U')
plt.plot(wn)
plt.show()
```

```python
def factorial(n):
  if n == 1:
    return 1

  return n * factorial(n - 1)

factorial(7)
```

```python
def binomial_rv(n, p):
  trues = 0
  for x in range(n):
    if np.random.uniform() < p:
      trues += 1
  return trues

binomial_rv(100, 0.4)
```

```python
def draw(k):
  payoff = 0
  count = 0

  for i in range(10):
    U = np.random.uniform()
    count = count + 1 if U < 0.5 else 0
    print(count)
    if count == k:
      payoff = 1

  return payoff


draw(2)
```

```python
from numpy.random import uniform

def draw_new(k):

    payoff = 0
    count = 0

    for i in range(10):
        U = uniform()
        count = count + ( 1 if U < 0.5 else 0 )
        print(count)
        if count == k:
            payoff = 1

    return payoff

draw_new(3)
```

```python
def fibonachi(x):
  if x == 1:
    return 1

  return x + fibonachi(x - 1)

fibonachi(3)
```
